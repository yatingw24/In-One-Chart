```python
import pandas as pd
```

```python
df = pd.read_csv('original_sorted.csv')

df
```

```python
df = df.drop(columns=['AVERAGE of sum'])

df
```

```python
df_long = df.melt(id_vars=["month"], 
                  var_name="category", 
                  value_name="average_value")

df_long["category"] = df_long["category"].str.replace("AVERAGE of ", "", regex=False)

print(df_long)
```

```python
df_long['average_value'] = df_long['average_value'] / 100
```

```python
df_long
```

```python
df_long.to_csv('output.csv', index=False)
```
